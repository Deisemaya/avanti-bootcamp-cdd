```python
# importando as bibliotecas
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df2 = sns.load_dataset('penguins')
df3 = sns.load_dataset('tips')
```

```python
df.head()
```

```python
# dividindo o valor de 1M em 1M
df = df.assign(arrecadacao_milhao = df.arrecadacao /1000000)
# ordenando as barras
df = df.sort_values(by='arrecadacao', ascending=False)
ax = sns.set_style('whitegrid')
plt.figure(figsize=(5, 2))
ax = sns.barplot(data=df, y='arrecadacao', x='titulo', color='tomato')
```

```python
#Adicionando valores nas barras
df = df.assign(arrecadacao_milhao = df.arrecadacao /1000000)
df = df.sort_values(by='arrecadacao', ascending=False)
ax = sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='tomato')

```

```python
#melhorando o titulo e o labels
df = df.assign(arrecadacao_milhao = df.arrecadacao /1000000)
df = df.sort_values(by='arrecadacao', ascending=False)
ax = sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='tomato')

plt.xlabel('Arrecadao (Milhes de dlares)')
plt.ylabel('Filme')
plt.title('Arrecadao de Filmes')


```

```python
#adicionando linhas verticais e removendo bordas
df = df.assign(arrecadacao_milhao = df.arrecadacao /1000000)
df = df.sort_values(by='arrecadacao', ascending=False)
ax = sns.set_style('whitegrid')
plt.figure(figsize=(5, 2))
ax = sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='tomato')


ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type='center',
    color='white'
)

plt.xlabel('Arrecadao (Milhes de dlares)')
plt.ylabel('Filme')
plt.title('Arrecadao de Filmes')

ax.set(frame_on=False)
plt.show()

```
